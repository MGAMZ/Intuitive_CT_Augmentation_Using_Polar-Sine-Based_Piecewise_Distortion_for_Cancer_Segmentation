步骤一：只保留backbone
步骤二：对齐backbone中各个参数名

```python
import torch

read_ckpt = r"D:/PostGraduate/DL/mgam_CT/mmpretrain/work_dirs/ViTSmall_MoCoV3_lr1e-4/epoch_300.pth"
save_ckpt = r"D:/PostGraduate/DL/mgam_CT/mmpretrain/work_dirs/ViTSmall_MoCoV3_lr1e-4/epoch_300_forMMseg.pth"
model = torch.load(read_ckpt)

print(model['state_dict'].keys())

# 把model['state_dict']中所有以‘backbone’开头的键，全部删除backbone
for k in list(model['state_dict'].keys()):
	if k.startswith('backbone'):
		model['state_dict'][k.replace('backbone.', '')] = model['state_dict'][k]

	del model['state_dict'][k]

print(model['state_dict'].keys())

# 键名命名规则对齐
for k in list(model['state_dict'].keys()):
	if 'attn.qkv.weight' in k:
		model['state_dict'][k.replace('attn.qkv.weight', 'attn.attn.in_proj_weight')] = model['state_dict'][k]
	elif 'attn.qkv.bias' in k:
		model['state_dict'][k.replace('attn.qkv.bias', 'attn.attn.in_proj_bias')] = model['state_dict'][k]
	elif 'attn.proj.weight' in k:
		model['state_dict'][k.replace('attn.proj.weight', 'attn.attn.out_proj.weight')] = model['state_dict'][k]
	elif 'attn.proj.bias' in k:
		model['state_dict'][k.replace('attn.proj.bias', 'attn.attn.out_proj.bias')] = model['state_dict'][k]
	else:
		continue
	del model['state_dict'][k]

print(model['state_dict'].keys())


# 保存新的模型
torch.save(model, save_ckpt)
```
